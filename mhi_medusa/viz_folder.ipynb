```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import os
from tqdm.contrib import tzip
```

```python
image_dir = r"E:\datasets\ck_master_dataset\png_960\a13_2022-10-28-07-00-04\l"
mhi_dir = r"E:\datasets\ck_master_dataset\mhi\a13_2022-10-28-07-00-04"
save_dir = r"E:\datasets\ck_master_dataset\mhi-detr_predict\a13_2022-10-28-07-00-04"
checkpoint = r"logs\MEDUSA-resnet50-batch-4-epoch-100\checkpoint.pth"
```

```python
mhi_paths = list(Path(mhi_dir).glob("*.npy"))
image_paths=[]
for mhi in mhi_paths:
    # print(mhi.stem[-8:])
    mhi_target = mhi.stem[-8:]+".png"
    image_path = Path(image_dir) / mhi_target
    if image_path.exists():
        image_paths.append(image_path)
assert len(mhi_paths) == len(image_paths)
print(mhi_paths[:5],image_paths[:5])
```

```python
for image_path,mhi_path in tzip(image_paths,mhi_paths):
    print(f"dealing pair: {image_path.name}, {mhi_path.name}")
    command = f"python main.py --image {str(image_path.absolute())} --mhi {str(mhi_path.absolute())} --resume {checkpoint} --dataset_file vitality --save_dir {save_dir}"
    # print(command)
    os.system(command)
    pass
```
